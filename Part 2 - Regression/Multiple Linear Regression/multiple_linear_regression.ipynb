## We will perform Backward Elimination

```python
# importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing the dataset
dataset = pd.read_csv('50_Startups.csv')
dataset
```

```python
X = dataset.iloc[:,:-1].values
X
```

```python
y = dataset.iloc[:,-1].values
y
```

```python
# Encoding the categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder = LabelEncoder()
X[:,3] = labelencoder.fit_transform(X[:,3])
X
```

```python
onehotencoder = OneHotEncoder(categorical_features= [3])
X = onehotencoder.fit_transform(X).toarray()
X
```

***
### Avoiding the dummy variable trap

![title](img/dummy_variable_trap.png)
***

```python
# Although we don't have to do it manually, libraries will do it automatically
X = X[:,1:]
X
```

```python
# splitting the dataset into the training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Fitting Multiple Liear Regression to the training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predicting the test set results
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

***
__Do you think it's actually the optimal model that we built with the dataset that we have here?  
You know because when we built this model we actually used all the independent variables.  
But what is among these independent variables there are some that are highly statistically significant  
that is have great impact on the dependent variable profit and some that are not statistically significant at all.  
That means that if we removed this non statistically significant variables from the modal  we would still get some amazing predictions__


***
![title](img/backward_elimination.png)
***

```python
# Building the optimal model using Backward Elimination
import statsmodels.formula.api as sm
X = np.append(arr = np.ones((50, 1)).astype(int), values = X, axis = 1)
X_opt = X[:, [0,1,2,3,4,5]]
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```

```python
#removing those features whose p value is more than significance level one by one 
X_opt = X[:, [0,1,3,4,5]]
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```

```python
# now removing x1
X_opt = X[:, [0,3,4,5]]
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```

```python
# removing the x2 feature from our previous set of X_opt that is X[:, 4] in our real matrix of features 
X_opt = X[:, [0, 3, 5]]
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```

```python
# removing the x2 feature from our previous set of X_opt that is  X[:, 5] in our real matrix of features 
X_opt = X[:, [0, 3]]
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```
